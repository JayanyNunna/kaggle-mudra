```python
import os
import shutil
import random
from tqdm import tqdm

# ✅ Updated source path to point to folder with class folders
source_path = "/kaggle/input/kuchipudi-mudra-dataset-master/Kuchipudi-Mudra-Dataset-master"
target_path = "/kaggle/working/restructured_dataset"

# Split ratios
split_ratio = [0.7, 0.15, 0.15]  # train, val, test
splits = ['train', 'val', 'test']

# Create target folders
for split in splits:
    os.makedirs(os.path.join(target_path, split), exist_ok=True)

# Process each class folder
for class_name in os.listdir(source_path):
    class_path = os.path.join(source_path, class_name)
    if not os.path.isdir(class_path):
        continue

    images = [f for f in os.listdir(class_path) if f.lower().endswith(('.jpg', '.jpeg', '.png'))]
    random.shuffle(images)

    n_total = len(images)
    n_train = int(split_ratio[0] * n_total)
    n_val = int(split_ratio[1] * n_total)

    split_files = {
        'train': images[:n_train],
        'val': images[n_train:n_train + n_val],
        'test': images[n_train + n_val:]
    }

    for split in splits:
        split_dir = os.path.join(target_path, split, class_name)
        os.makedirs(split_dir, exist_ok=True)
        for img_name in split_files[split]:
            shutil.copy2(os.path.join(class_path, img_name), os.path.join(split_dir, img_name))

print("✅ Dataset restructured into train/val/test.")

```

```python
import os
import shutil

base_dir = '/kaggle/working/restructured_dataset'

# Loop over train, val, test folders
for split in ['train', 'val', 'test']:
    split_dir = os.path.join(base_dir, split)
    contents = os.listdir(split_dir)

    if len(contents) == 1:
        inner = os.path.join(split_dir, contents[0])
        # If there's only one folder inside, and it's the .zip root folder
        if os.path.isdir(inner):
            for item in os.listdir(inner):
                src = os.path.join(inner, item)
                dst = os.path.join(split_dir, item)
                shutil.move(src, dst)
            shutil.rmtree(inner)
            print(f"✅ Fixed folder structure for {split}")
        else:
            print(f"⚠️ Unexpected file inside {split}")
    else:
        print(f"✅ {split} already seems fine")

```

```python
from tensorflow.keras.preprocessing.image import ImageDataGenerator

# Set up data generators
train_datagen = ImageDataGenerator(
    rescale=1./255,
    rotation_range=20,
    shear_range=0.4,
    zoom_range=0.2,
    horizontal_flip=True
)

val_test_datagen = ImageDataGenerator(rescale=1./255)

# Flow from directories
train_generator = train_datagen.flow_from_directory(
    os.path.join(target_path, 'train'),
    target_size=(128, 128),
    batch_size=32,
    class_mode='categorical',
    shuffle=True
)

val_generator = val_test_datagen.flow_from_directory(
    os.path.join(target_path, 'val'),
    target_size=(128, 128),
    batch_size=32,
    class_mode='categorical',
    shuffle=False
)

test_generator = val_test_datagen.flow_from_directory(
    os.path.join(target_path, 'test'),
    target_size=(128, 128),
    batch_size=32,
    class_mode='categorical',
    shuffle=False
)

print("✅ Data generators ready.")
```

```python
from tensorflow.keras.applications import VGG16
from tensorflow.keras.models import Model

# Step 1: Create base model without weights
base_model = VGG16(weights=None, include_top=False, input_shape=(128, 128, 3))

# Step 2: Load weights manually
base_model.load_weights('/kaggle/input/vgg16-weights-tf-dim-ordering-tf-kernels-notop-h5/vgg16_weights_tf_dim_ordering_tf_kernels_notop.h5')

# Step 3: Freeze layers
for layer in base_model.layers:
    layer.trainable = False

# Step 4: Define feature extractor
feature_extractor = Model(inputs=base_model.input, outputs=base_model.get_layer('block5_pool').output)

print("✅ VGG16 feature extractor ready.")

```

```python
import numpy as np
from tensorflow.keras.applications.vgg16 import preprocess_input

# Helper to extract features
def extract_features(generator, feature_extractor):
    features = []
    labels = []
    for i in tqdm(range(len(generator))):
        x_batch, y_batch = generator[i]
        x_batch = preprocess_input(x_batch)
        feature_maps = feature_extractor.predict(x_batch)
        flat_features = feature_maps.reshape((feature_maps.shape[0], -1))
        features.append(flat_features)
        labels.append(y_batch)
    X = np.vstack(features)
    y = np.vstack(labels)
    return X, y
```

```python
X_train, y_train = extract_features(train_generator, feature_extractor)
X_val, y_val = extract_features(val_generator, feature_extractor)
X_test, y_test = extract_features(test_generator, feature_extractor)

print("✅ Feature extraction complete.")
print("Train shape:", X_train.shape, y_train.shape)
print("Val shape:", X_val.shape, y_val.shape)
print("Test shape:", X_test.shape, y_test.shape)
```
